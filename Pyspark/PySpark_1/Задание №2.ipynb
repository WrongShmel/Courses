```python
from pyspark.sql import SparkSession
from pyspark import SparkConf, SparkContext
from datetime import datetime
import re
```

```python
conf=SparkConf().setAppName("Number2").setMaster("local[*]")
sc=SparkContext(conf=conf)
```

```python
read_line=sc.textFile("/home/jovyan/3_5_Independent_work/password.txt")

clear_pass=(read_line.map(lambda x: x.strip())
                     .filter(lambda x: x != "")
)

print(clear_pass.take(5))
```

```python
light_rdd=clear_pass.map(lambda x: len(x))

print(light_rdd.take(5))

avg_pass=(light_rdd.map(lambda x: (x, 1))
                   .reduce(lambda x, y: (x[0]+y[0], x[1]+y[1]))
)
avg=int(round(avg_pass[0]/avg_pass[1], 0))

max_pass=(light_rdd.reduce(lambda x, y: x if x > y else y)
)

min_pass=(light_rdd.reduce(lambda x, y: x if x < y else y)
)

top_pass=(light_rdd.map(lambda x: (x, 1))
                   .reduceByKey(lambda x, y: x+y)
                   .top(5, key=lambda x: x[1])
)
```

```python
only_number=(clear_pass.filter(lambda x: x.isdigit())
                       .count()
)

only_letters=(clear_pass.filter(lambda x: x.isalpha())
                        .count()
              )
```

```python
prefix_pass=(clear_pass.map(lambda x: (x[0:3], 1))
                       .reduceByKey(lambda x,y: x+y)
                       .top(5, key=lambda x: x[1])
             )

suffix_pass=(clear_pass.map(lambda x: (x[-3:], 1))
                       .reduceByKey(lambda x,y: x+y)
                       .top(5, key=lambda x: x[1])
             )
```

```python
print(f"Средняя длина пароля: {avg}")
print(f"Минимальная длина пароля {min_pass}")
print(f"Максимальная длина пароля {max_pass}")
print(*top_pass, sep="\n")

print(only_number)
print(only_letters)

print(*prefix_pass, sep="\n")
print(*suffix_pass, sep="\n")
```
