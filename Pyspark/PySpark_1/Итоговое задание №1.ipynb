```python
from pyspark.sql import SparkSession
from pyspark import SparkConf, SparkContext
from datetime import datetime
```

```python
conf = SparkConf().setAppName("Itog").setMaster("local[*]")
sc = SparkContext(conf=conf)
```

```python
countries_rdd=sc.textFile("/home/jovyan/3_5_Independent_work/List of Countries.txt")
```

```python
countries=(countries_rdd
                        .map(lambda x: (x[0], 1))
                        .reduceByKey(lambda x, y: x + y)
                        .sortBy(lambda x: x[0], ascending=True)
                       #.sortByKey()
          )
```

```python
lengths_countries=(countries_rdd
                                .map(lambda x: (x, len(x)))
                     )
```

```python
top_5_long_countries=(lengths_countries
                        .sortBy(lambda x: x[1], ascending=True)
                        .top(5, key=lambda x: [0])
                      )
```

```python
top_5_short_countries=(lengths_countries
                       .sortBy(lambda x: x[1], ascending=False)
                       .top(5, key=lambda x: [0])
                      )
```

```python
avg_countries=(lengths_countries
               .map(lambda x: x[1])
               .reduce(lambda x, y: x + y)
              )/lengths_countries.count()
```

```python
two_words=(countries_rdd.filter(lambda x: x if " " in x else ""))
```

```python
print("ОТВЕТ НА ПЕРВОЕ ЗАДАНИЕ", "\n")
print(*countries.collect(), "\n", sep="\n")
print("ОТВЕТ НА ВТОРОЕ ЗАДАНИЕ", "\n")
print(*top_5_long_countries, "\n", sep="\n")
print("ОТВЕТ НА ТРЕТЬЕ ЗАДАНИЕ", "\n")
print(*top_5_short_countries, "\n", sep="\n")
print("ОТВЕТ НА ЧЕТВЕРТОЕ ЗАДАНИЕ", "\n")
print(round(avg_countries, 1), "\n", sep="\n")
print("ОТВЕТ НА ПЯТОЕ ЗАДАНИЕ", "\n")
print(*two_words.collect(), "\n", sep="\n")
```
