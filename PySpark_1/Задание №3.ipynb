```python
from pyspark.sql import SparkSession
from pyspark import SparkConf, SparkContext
from datetime import datetime
import re
```

```python
conf=SparkConf().setAppName("Number3").setMaster("local[*]")
sc=SparkContext(conf=conf)
```

```python
read_dict=sc.textFile("/home/jovyan/3_5_Independent_work/Rus_dict.txt")

clear_dict=(read_dict.map(lambda x: x.strip())
                     .filter(lambda x: x != "")
                     .filter(lambda x: x.strip())
                     .distinct()
)
```

```python
count_words=clear_dict.count()
long_words=(clear_dict.map(lambda x: (x, len(x)))
                      .top(5, key=lambda x: x[1])
            )
short_words=(clear_dict.map(lambda x: (x, len(x)))
                      .top(5, key=lambda x: -x[1])
            )

all_short_words=(clear_dict.map(lambda x: (x, len(x)))
                      .filter(lambda x: x[1] == 2)
            )
```

```python
count_words_length=(clear_dict.map(lambda x: (len(x), 1))
                              .reduceByKey(lambda x, y: x+y)
                              .sortByKey()
)
words_length=(count_words_length.top(1, key=lambda x: x[1]))
```

```python
count_letter=(clear_dict.filter(lambda x: x if "ё" in x else 0)
)
palindrome=(clear_dict.filter(lambda x: x if x == x[::-1] else 0)
)
```

```python
print(f"Общее количество слов в словаре: {count_words}")
print(f"Топ-5 самых длинных слов")
for i in long_words:
  print(f"{i[0]}: {i[1]}")
print(f"Топ-5 самых коротких слов")
for i in short_words:
  print(f"{i[0]}: {i[1]}")

for i in count_words_length.collect():
  print(f"{i[0]}: {i[1]}")
print(f"Длина слова, которая встречается чаще всего: {words_length[0][0]}")

print(f"Количество слов содержащих букву ё: {count_letter.count()}")
for i in palindrome.collect():
  print(f"{i}")
```
